```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
cities = pd.read_csv(r'C:\Users\wreny\Desktop\Data Programming Class\NCSC_zip_2.csv')

    
map = Basemap(projection='merc', 
    lat_0 = 35.630066, lon_0 = -79.806419,# Map is centered at this point, North Carolina LAT and LONG
    resolution = 'f', area_thresh = 0.1,
    llcrnrlon=-84.15527343750001, llcrnrlat=31.541089879585837,#lowerleft coordinates of map
    urcrnrlon=-75.01464843750001, urcrnrlat=36.87962060502676) #Upperright coordinates of map

lat = cities['LAT'].values #stores Lat values in list
lon = cities['LNG'].values #stores Long values in list
impact = cities['IMPACT'].values #stores Impact level in list (i.e, high Medium Low)

#for each level of impact iterated through in the stored list the corresponding
#lat and Long values are plotted and given a corresponding marker and color
for x in range (len(impact)):
    if impact[x] == 'HIGH':
        x,y = map(lon[x], lat[x])
        map.plot(x, y, 'ro', markersize=12) 
    elif impact[x] == 'LOW':
        x,y = map(lon[x], lat[x])
        map.plot(x, y, 'go', markersize=12)
    elif impact[x] == 'MEDIUM':
        x,y = map(lon[x], lat[x])
        map.plot(x, y, 'bo', markersize=12)
            
#draws county lines with several line caracteristics
map.drawcounties(linewidth=1, 
                 linestyle='solid', 
                 color='k', 
                 antialiased=1, 
                 ax=None, 
                 zorder=None, 
                 drawbounds=False)
map.drawstates()
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color = 'coral')
map.drawmapboundary()


plt.rcParams['figure.figsize'] = (50,50) #sets ize of map
plt.show()
```
