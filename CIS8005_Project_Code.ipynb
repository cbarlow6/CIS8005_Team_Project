```python
# Reading the file with the offset of latitude and longitudes, 48 hours prior to landfall
# and location of landfall, for various hurricanes after 2010
Location = r'Landfall.csv' 
df5 = pd.read_csv(Location)
df5

#Calculating mean latitude offset value 
latOffset = pd.Series(df5['Lat'].values).abs()
print latOffset.mean()

#Calculating mean latitude offset value
lngOffset = pd.Series(df5['Long'].values).abs()
print lngOffset.mean()
```

```python
# Calculating the distances of each of the zipcodes from the predicted location of the landfall,
# adding a column to the dataframe
df2['DIST'] = earthRad * np.arccos(np.sin(np.radians(df2['LAT'])) * np.sin(np.radians(refLat)) + np.cos(np.radians(df2['LAT'])) * np.cos(np.radians(refLat)) * np.cos((np.radians(df2['LNG']) - np.radians(refLng))))

df2.to_csv('NCSC_zip_2.csv', index = False)
df2
```

```python
# Reading the file with population by zipcodes in USA 
Location = r'population_by_zip_2010.csv' 
df3 = pd.read_csv(Location)

#Filtering the population data for the relevant zipcode
df3[(df3['gender'].isnull()) & (df3['zipcode'] > 27000) & (df3['zipcode'] < 30000)].to_csv('NCSC_pop_zip.csv', index = False)

#Reading the filtered population data
Location2 = r'NCSC_pop_zip.csv' 
df4 = pd.read_csv(Location2)


#Sorting the population data by zipcode and storing them in ascending order 
df4 = df4.sort_values(['zipcode'])
df4.to_csv('NCSC_pop_zip.csv', index = False)
```

```python
# Storing the zipcode and population in dictionary
d = df4.set_index('zipcode')['population'].to_dict()

# Adding a population column to our main dataframe using the dictionary
df2['POPULATION'] = df2['ZIP'].map(d)
df2.to_csv('NCSC_zip_2.csv', index = False)
df2
```

```python
# According to USA TODAY, the radius of the hurricance force winds is approx. 120 km
# Adding a column to define the IMPACT ZONE of a zipcode
# HIGH Impact - within a 120 km radius
# MEDIUM Impact - in a radius above 120 km and below 250 km 
# LOW Impact - anything in a radius above 250 km

df2['IMPACT'] = df2['DIST'].map(lambda x: 'HIGH' if x < 120 else ('MEDIUM' if 120 < x < 250 else 'LOW'))

# The population density is defined as follows
# HIGH Population - greater than 25000
# MEDIUM Population - between 10000 and 25000
# LOW Population - less than 10000

df2['POP_DENSITY'] = df2['POPULATION'].map(lambda x: 'LOW' if x < 10000 else ('MEDIUM' if 10000 < x < 25000 else 'HIGH'))

df2.to_csv('NCSC_zip_2.csv',index = False)
df2
```

```python
# According to past hurricane data and expert opinion 
# On an average around 100 zipcodes suffer power cuts, which are divided as follows
# Approx. 60 in the HIGH impact zones 
# Approx. 30 in the MEDIUM impact zones 
# Approx. 10 in the LOW impact zones 

#Creating a subset doe each type of impact with zipcodes that are likely to suffer power cuts
max_subset = df2[df2['IMPACT']=='HIGH'].sample(n=60)
max_subset = max_subset.sort_values(['POPULATION'])
max_subset
popAffect = max_subset['POPULATION'].sum()

med_subset = df2[df2['IMPACT']=='MEDIUM'].sample(n=30)
med_subset = med_subset.sort_values(['POPULATION'])
med_subset
popAffect += med_subset['POPULATION'].sum()

low_subset = df2[df2['IMPACT']=='LOW'].sample(n=10)
low_subset = low_subset.sort_values(['POPULATION'])
low_subset
popAffect += low_subset['POPULATION'].sum()
```
